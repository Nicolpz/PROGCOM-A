#16) ReferenceError
Error generado porque el objeto referido ya no existe

```python
import gc
import weakref
class Foo(object):
    def __init__(self, name):
        self.name = name
    def __del__(self):
        print ('(Deleting %s)' % self)
obj = Foo('obj')
p = weakref.proxy(obj)
print ('BEFORE:', p.name)
obj = None
print ('AFTER:', p.name)
```

#19) AssertionError
Error que nos indica que algo imposible sucedió

```python
price=20
discount= 60 
def calculate_discount(price, discount):
    discounted_price = price - (discount*price)
    assert 0 <= discounted_price <= price
    print(discounted_price)
    return;
calculate_discount(price, discount)
```
